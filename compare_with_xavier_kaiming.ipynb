```python
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import numpy as np

def plot_dist(a, b, c):
    fig, ax = plt.subplots(1, 3, figsize=(18,6))
    ax[0].hist(a, bins = 100, label='a', density=True)
    ax[0].hist(b, bins = 100, label='b', density=True)
    ax[0].hist(c, bins = 100, label='c', density=True)
    ax[0].set_title('original')
    ax[0].legend(title='distribution', shadow=True, labelspacing=0.5)

    a_x = nn.init.xavier_normal_(torch.tensor(a))
    b_x = nn.init.xavier_normal_(torch.tensor(b))
    c_x = nn.init.xavier_normal_(torch.tensor(c))
    ax[1].hist(a_x.numpy(), bins = 100, density=True)
    ax[1].hist(b_x.numpy(), bins = 100, density=True)
    ax[1].hist(c_x.numpy(), bins = 100, density=True)
    ax[1].set_title('xavier_normal_')

    a_k = nn.init.kaiming_normal_(torch.tensor(a))
    b_k = nn.init.kaiming_normal_(torch.tensor(b))
    c_k = nn.init.kaiming_normal_(torch.tensor(c))
    ax[2].hist(a_k.numpy(), bins = 100, density=True)
    ax[2].hist(b_k.numpy(), bins = 100, density=True)
    ax[2].hist(c_k.numpy(), bins = 100, density=True)
    ax[2].set_title('kaiming_normal_')

n = 10000
a = 3 * np.random.randn(n,1) - 2
b = np.random.randn(n,1)
c = 1/3 * np.random.randn(n,1) + 2
plot_dist(a,b,c)

a = 3 * np.random.rand(n,1) - 2
b = np.random.rand(n,1)
c = 1/3 * np.random.rand(n,1) + 2
plot_dist(a,b,c)
```
