```python
import numpy as np
import scipy.stats as stats
from matplotlib import pyplot as plt 

s = 1
e = 18 + 1
size = 300

data = np.random.randint(s, e, size)
#data = np.random.uniform(s, e, size)

plt.hist(data) 
plt.title("histogram") 
plt.show()
```

```python
import seaborn as sns

sns.distplot(data, hist = True, label='data')
stats.shapiro(data)
```

```python
type(data)

m2 = []
ld = data.tolist()
count = 3000
sample_size = 100

import random

for n in range(1, count):
    data2 = random.sample(ld, sample_size)
    #data2 = np.random.randint(e, size=n)
    m2.append(np.mean(data2))

plt.hist(m2)
plt.title("mean histogram")
plt.show()
```

```python
sns.distplot(m2).set_title('Probability distribution')
#len(m2)
print("확률분포 : ", stats.shapiro(m2)[1])
```

```python
ma2 =np.asarray(m2)

z = (ma2 - ma2.mean()) / ma2.std()
#z = (ma2.mean() - 9)/()

sns.distplot(z).set_title('Z distribution')
print("z 분포 : ", stats.shapiro(z)[1])
#len(m2)
```

```python
sns.distplot(z)
x = np.random.normal(size=100000)
sns.distplot(x)
plt.axvline(1.65, linestyle='--', color='red', alpha=.7)
print("표준정규분포 : ", stats.shapiro(x)[1])
```
